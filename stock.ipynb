```python
import numpy as np
import pandas as pd
import datetime

# Fungsi untuk konversi string ke datetime
def str_to_datetime(s):
    split = s.split('-')
    year, month, day = int(split[0]), int(split[1]), int(split[2])
    return datetime.datetime(year=year, month=month, day=day)

# Fungsi utama untuk membentuk windowed dataframe
def df_to_windowed_df(dataframe, first_date_str, last_date_str, n=3):
    first_date = str_to_datetime(first_date_str)
    last_date = str_to_datetime(last_date_str)

    target_date = first_date
    dates = []
    X, Y = [], []

    while True:
        # Ambil subset data hingga tanggal target
        df_subset = dataframe.loc[:target_date].tail(n+1)

        # Jika data tidak mencukupi, isi kekurangan dengan interpolasi atau default
        if len(df_subset) < n+1:
            missing_count = (n+1) - len(df_subset)
            print(f"Filling {missing_count} missing rows for date {target_date}.")
            filler = pd.DataFrame(
                np.nan,
                index=pd.date_range(end=target_date, periods=missing_count, freq='D'),
                columns=dataframe.columns
            )
            df_subset = pd.concat([filler, df_subset]).sort_index().interpolate(method='linear').fillna(method='bfill')

        # Ambil nilai X dan Y dari subset
        values = df_subset['Close'].to_numpy()
        x, y = values[:-1], values[-1]

        dates.append(target_date)
        X.append(x)
        Y.append(y)

        # Cari tanggal berikutnya (1 minggu kemudian)
        next_week = dataframe.loc[target_date:target_date + datetime.timedelta(days=7)]
        if next_week.empty:
            break

        next_datetime_str = str(next_week.head(2).tail(1).index.values[0])
        next_date_str = next_datetime_str.split('T')[0]
        year, month, day = map(int, next_date_str.split('-'))
        next_date = datetime.datetime(year=year, month=month, day=day)

        if next_date > last_date:
            break

        target_date = next_date

    # Buat DataFrame keluaran
    ret_df = pd.DataFrame({'Target Date': dates})

    X = np.array(X)
    for i in range(0, n):
        ret_df[f'Target-{n-i}'] = X[:, i]

    ret_df['Target'] = Y

    return ret_df

# Simulasi data saham GGRM
dates = pd.date_range(start="2019-07-01", end="2024-10-01", freq='D')  # Rentang tanggal
close_values = np.random.uniform(40000, 60000, size=len(dates))  # Simulasi harga penutupan dengan panjang sesuai

# Konversi ke DataFrame
data = {
    "Date": dates,
    "Close": close_values
}
df = pd.DataFrame(data)
df.set_index("Date", inplace=True)

# Tampilkan 5 baris pertama dataset
print("Dataset Simulasi GGRM:")
print(df.head())

# Jalankan fungsi untuk membuat windowed dataframe tanpa skipping date
windowed_df = df_to_windowed_df(df, '2019-07-29', '2024-10-01', n=3)

# Tampilkan hasil
print("\nWindowed DataFrame:")
print(windowed_df)
```
